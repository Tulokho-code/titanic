```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


train_df = pd.read_csv("data/train.csv")

sns.set(style="whitegrid", palette="muted")

# Tạo subplot để hiển thị nhiều biểu đồ
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# (1) Phân bố tuổi hành khách
sns.histplot(train_df['Age'].dropna(), bins=30, kde=True, color="skyblue", ax=axes[0,0])
axes[0,0].set_title("Age Distribution of Passengers")
axes[0,0].set_xlabel("Age")
axes[0,0].set_ylabel("Count")

# (2) Tỷ lệ nam/nữ
sns.countplot(x='Sex', data=train_df, palette="Set2", ax=axes[0,1])
axes[0,1].set_title("Gender Distribution")
axes[0,1].set_xlabel("Sex")
axes[0,1].set_ylabel("Count")

# (3) Phân bố Fare theo Class
sns.boxplot(x="Pclass", y="Fare", data=train_df, palette="pastel", ax=axes[1,0])
axes[1,0].set_title("Fare Distribution by Passenger Class")
axes[1,0].set_xlabel("Passenger Class")
axes[1,0].set_ylabel("Fare")

# (4) Tỷ lệ sống sót tổng
survival_rate = train_df['Survived'].mean() * 100
sns.countplot(x='Survived', data=train_df, palette="coolwarm", ax=axes[1,1])
axes[1,1].set_title(f"Overall Survival Rate ({survival_rate:.1f}%)")
axes[1,1].set_xlabel("Survived (0 = No, 1 = Yes)")
axes[1,1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```
